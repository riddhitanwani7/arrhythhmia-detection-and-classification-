```python
# 1. Load and Prepare Data
import pandas as pd
import numpy as np
import random
import zipfile
import os
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, mean_absolute_error, cohen_kappa_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns

from imblearn.over_sampling import SMOTE

# Set seed globally to the best found
BEST_SEED = 10
random.seed(BEST_SEED)
np.random.seed(BEST_SEED)

# Extract data from ZIP
zip_path = r"C:\\Users\\Riddhi\\Downloads\\arrhythmia.zip"
extract_path = "arrhythmia"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Load and clean data
df = pd.read_csv('arrhythmia/arrhythmia.data', header=None)
df = df.replace('?', pd.NA)
df = df.apply(pd.to_numeric, errors='coerce')
df.fillna(df.median(), inplace=True)

# Separate features and target
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Define models with fixed random_state
models = {
    'Random Forest': RandomForestClassifier(random_state=BEST_SEED),
    'SVM': SVC(kernel='linear', max_iter=1000, random_state=BEST_SEED),
    'KNN': KNeighborsClassifier(n_neighbors=3),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=BEST_SEED)
}

# Evaluation with 10-fold Stratified CV and result logging
def evaluate_with_cv(X, y, label):
    print(f"\n--- {label} ---")
    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=BEST_SEED)
    results = {name: {'Accuracy': [], 'MAE': [], 'Kappa': []} for name in models}

    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y)):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        for name, model in models.items():
            try:
                model.fit(X_train, y_train)
                preds = model.predict(X_test)

                acc = accuracy_score(y_test, preds)
                mae = mean_absolute_error(y_test, preds)
                kappa = cohen_kappa_score(y_test, preds)

                results[name]['Accuracy'].append(acc)
                results[name]['MAE'].append(mae)
                results[name]['Kappa'].append(kappa)
            except Exception as e:
                print(f"Error in {name} on fold {fold+1}: {e}")

    summary = []
    for name in models:
        acc_mean = np.mean(results[name]['Accuracy'])
        mae_mean = np.mean(results[name]['MAE'])
        kappa_mean = np.mean(results[name]['Kappa'])

        print(f"\n{name} (Averaged over 10 folds):")
        print("Accuracy:", acc_mean)
        print("MAE:", mae_mean)
        print("Kappa:", kappa_mean)

        summary.append({"Model": name, "Stage": label, "Accuracy": acc_mean, "MAE": mae_mean, "Kappa": kappa_mean})

    return pd.DataFrame(summary)

# Convert to arrays
X_array = X.values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_array)

# 1. Evaluate BEFORE SMOTE & Feature Selection
results_1 = evaluate_with_cv(X_scaled, y.values, label="Before SMOTE & Feature Selection")

# 2. SMOTE applied to entire dataset
smote = SMOTE(random_state=BEST_SEED, k_neighbors=1)
X_smote, y_smote = smote.fit_resample(X_scaled, y)

results_2 = evaluate_with_cv(X_smote, y_smote, label="After SMOTE Before Feature Selection")

# 3. Feature Selection: Lasso + RF-RFE
# K-part Lasso
def k_part_lasso(X, y, k=5):
    n_features = X.shape[1]
    part_size = n_features // k
    selected = []

    for i in range(k):
        X_part = X[:, i * part_size: (i + 1) * part_size]
        lasso = LassoCV(cv=5, random_state=BEST_SEED).fit(X_part, y)
        selected_idx = np.where(lasso.coef_ != 0)[0]
        selected.extend((i * part_size + selected_idx).tolist())

    X_selected = X[:, selected]
    lasso_final = LassoCV(cv=5, random_state=BEST_SEED).fit(X_selected, y)
    final_selected_idx = np.where(lasso_final.coef_ != 0)[0]

    return X_selected[:, final_selected_idx]

X_lasso = k_part_lasso(X_smote, y_smote)

# RFE
rf = RandomForestClassifier(n_estimators=100, random_state=BEST_SEED)
rfe = RFE(rf, n_features_to_select=89)
X_final = rfe.fit_transform(X_lasso, y_smote)

results_3 = evaluate_with_cv(X_final, y_smote, label="After Feature Selection")

# Combine and save all results
all_results = pd.concat([results_1, results_2, results_3], ignore_index=True)
all_results.to_csv("arrhythmia_model_results.csv", index=False)

# Plot accuracy comparison
plt.figure(figsize=(10, 6))
sns.barplot(data=all_results, x="Model", y="Accuracy", hue="Stage")
plt.title("Accuracy Comparison Across Models and Stages")
plt.ylabel("Mean Accuracy")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
